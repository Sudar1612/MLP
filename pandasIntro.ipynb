# Groupby
Three Stages
* Split - split dataframe into multiple smaller dataframe based on the values of keys
* Apply - apply desired aggregation / transformation on each dataframe
* Combine - combine results from apply state into a dataframe

```python
df=pd.DataFrame({
    'key' : ['A','B','C']*2,  #can also be wrriten as list("ABCABC")
    'data':range(6)
})
df
```

```python
df.groupby("key").sum()
```
